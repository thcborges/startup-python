```python
import string
from random import choices
from re import sub

from pyspark.sql import SparkSession
```

```python
app_hash = "".join(choices(string.ascii_letters + string.digits, k=12))

spark = (
    SparkSession.builder.appName(app_hash)
    .master("local[*]")
    .enableHiveSupport()
    .getOrCreate()
)
```
